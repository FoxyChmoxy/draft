```python
df = pd.read_excel('reviews.xlsx')
df = df[df['review_text'].notnull() & df['author_name'].notnull()]
df["review_subtitle"] = df["review_subtitle"].replace([float('nan')] , ["Пусто"])
```

```python
df['s'] = df['review_text'].astype('str')
```

```python
df['s'].replace('[^0-9а-яА-ЯёЁ]+',' ',regex=True, inplace = True)
df['s']
```

```python
def fresh(sentence):
    words = [word.lower().strip() for word in sentence.split()]
    return ' '.join(words)

df['s'] = df['s'].apply(fresh)
df['s']
```

```python
df['t'] = df['s'].apply(lambda y: y.split())
df['t']
```
